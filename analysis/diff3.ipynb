## Visão por arquivos

```python
files_results_query = f"""
    SELECT
        ge.language,
        ge.tool,
        ge.result,
        COUNT(ge.file_path) as total_files
    FROM
        global_executions ge
    WHERE
        EXISTS (SELECT 1 FROM global_executions_per_commit gepc where gepc.scenario_id = ge.scenario_id)
    GROUP BY
        ge.tool,
        ge.result
    ORDER BY
        ge.language,
        ge.tool,
        ge.result
"""
sql_query_to_pd(files_results_query)
```

## Visão por commits

```python
scenarios_results_query = f"""
    WITH total_of_scenarios AS (
        SELECT
            COUNT(DISTINCT gepc.scenario_id) as total_of_scenarios
        FROM 
            global_executions_per_commit gepc
    )
    SELECT
        gepc.language,
        gepc.tool,
        gepc.result,
        COUNT(gepc.scenario_id) as scenarios,
        ROUND(COUNT(gepc.scenario_id) * 100. / (SELECT total_of_scenarios FROM total_of_scenarios), 1)
    FROM
        global_executions_per_commit gepc
    GROUP BY
        gepc.language,
        gepc.tool,
        gepc.result
    ORDER BY
        gepc.language,
        gepc.tool,
        gepc.result
"""
sql_query_to_pd(scenarios_results_query)
```
